```python
import cv2 as cv
import dlib
import matplotlib.pyplot as plt
```

```python
img = cv.imread("data/fotos/grupo.0.jpg")
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
```

```python
detector = dlib.cnn_face_detection_model_v1("data/recursos/mmod_human_face_detector.dat")
detected_faces = detector(img, 1)

img_with_squares = img.copy()

for detected_face in detected_faces:
    confidence = detected_face.confidence
    rect = detected_face.rect

    x_min, y_max = rect.tl_corner().x, rect.tl_corner().y
    x_max, y_min = rect.br_corner().x, rect.br_corner().y

    cv.rectangle(img_with_squares, (x_min, y_max), (x_max, y_min), (0,0,255), 1)
    cv.putText(img_with_squares, f"{confidence:.2f}", (rect.bl_corner().x,rect.bl_corner().y + 40), cv.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 2, cv.LINE_AA)


plt.imshow(cv.cvtColor(img_with_squares, cv.COLOR_BGR2RGB))
```
