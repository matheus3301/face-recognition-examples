```python
import cv2 as cv
import dlib
import matplotlib.pyplot as plt
```

## Criando detector de faces e de pontos

```python
face_detector = dlib.get_frontal_face_detector()

# detector de 68 pontos na face
face_points_detector = dlib.shape_predictor("data/recursos/shape_predictor_68_face_landmarks.dat")
```

## Imagem de Teste

```python
img = cv.imread("data/fotos/treinamento/ronald.0.0.jpg")

faces = face_detector(img, 2)

for face in faces:
    detected_points = face_points_detector(img, face)
    
    for i in range(detected_points.num_parts):
        (x, y) = detected_points.part(i).x, detected_points.part(i).y
        cv.circle(img, (x,y), 1, (0,0,255), 1, cv.LINE_AA)

plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
```
