```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
classifier = cv.CascadeClassifier("data/recursos/haarcascade_frontalface_default.xml")
```

```python
img = cv.imread("data/fotos/grupo.0.jpg")
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
```

```python
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
detected_faces = classifier.detectMultiScale(img_gray, scaleFactor=1.2, minSize=(60,60))

img_with_squares = img.copy()

for detected_face in detected_faces:
    x = detected_face[0]
    y = detected_face[1]
    width = detected_face[2]
    height = detected_face[3]

    cv.rectangle(img_with_squares, (x,y), (x + width, y + height), (0,0,255), 1)

plt.imshow(cv.cvtColor(img_with_squares, cv.COLOR_BGR2RGB))
```
